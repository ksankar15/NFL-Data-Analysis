```python
import numpy as np
import pandas as pd
receiving_2019 = pd.read_csv('Receiving_2019.csv')
receiving_2020 = pd.read_csv('Receiving_2020.csv')
receiving_2021 = pd.read_csv('Receiving_2021.csv')
receiving_2022 = pd.read_csv('Receiving_2022.csv')
receiving_2023 = pd.read_csv('Receiving_2023.csv')
receiving_2024 = pd.read_csv('Receiving_2024.csv')
```

```python
total_receiving = pd.concat([receiving_2019, receiving_2020, receiving_2021, receiving_2022, receiving_2023, receiving_2024], axis=0)
total_receiving
```

```python
total_receiving.to_csv("Total_Receiving.csv", index = False)
print("Data saved to Total_Receiving.csv")
```
