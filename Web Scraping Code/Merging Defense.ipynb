```python
import numpy as np
import pandas as pd
defense_2019 = pd.read_csv('Defense_2019.csv')
defense_2020 = pd.read_csv('Defense_2020.csv')
defense_2021 = pd.read_csv('Defense_2021.csv')
defense_2022 = pd.read_csv('Defense_2022.csv')
defense_2023 = pd.read_csv('Defense_2023.csv')
defense_2024 = pd.read_csv('Defense_2024.csv')
```

```python
total_defense = pd.concat([defense_2019, defense_2020, defense_2021, defense_2022, defense_2023, defense_2024], axis=0)
total_defense
```

```python
total_defense.to_csv("Total_Defense.csv", index = False)
print("Data saved to Total_Defense.csv")
```
