```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.nflpenalties.com/all-players.php?view=reg_season&year=2024'

# Call the GET request
requests = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(requests.text, "html.parser")

# examined HTML output using inspect to find where data is
columns = ['Name', 'Position', 'Team', 'Penalty Count', 'Yards', 'Declined', 'Offsetting', 'Total Flags', 'Pre-Snap', 'Count per Game', 'Yards per Game', 'Percent of Team', 'Penalty Types']
player_penalties_df_2024 = pd.DataFrame(columns = columns)

penalties_div = soup.find(name = 'div', attrs = {'class' : 'inner'})
table = penalties_div.find("table") 
tbody = table.find("tbody")  # Get the tbody element
rows = tbody.find_all("tr")  # Find all table rows

for row in rows:
    cols = row.find_all("td")  # Get all table data cells
    data = [col.text.strip() for col in cols]  # Extract text from each column
    penalty_series = pd.Series(data, index = columns)
    player_penalties_df_2024 = pd.concat([player_penalties_df_2024, penalty_series.to_frame().T], ignore_index=True)

display(player_penalties_df_2024)

player_penalties_df_2024.to_csv("2024_Player_Penalties.csv", index = False)
print("Data saved to 2024_Player_Penalties.csv")
```
