```python
import numpy as np
import pandas as pd
rushing_2019 = pd.read_csv('Rushing_2019.csv')
rushing_2020 = pd.read_csv('Rushing_2020.csv')
rushing_2021 = pd.read_csv('Rushing_2021.csv')
rushing_2022 = pd.read_csv('Rushing_2022.csv')
rushing_2023 = pd.read_csv('Rushing_2023.csv')
rushing_2024 = pd.read_csv('Rushing_2024.csv')
```

```python
total_rushing = pd.concat([rushing_2019, rushing_2020, rushing_2021, rushing_2022, rushing_2023, rushing_2024], axis=0)
total_rushing.rename(columns = {'Rushign Touchdowns': 'Rushing Touchdowns'}, inplace = True)
total_rushing
```

```python
total_rushing.to_csv("Total_Rushing.csv", index = False)
print("Data saved to Total_Rushing.csv")
```
