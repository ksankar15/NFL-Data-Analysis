```python
import numpy as np
import pandas as pd
penalty_2019 = pd.read_csv('Player_Penalties_2019.csv')
penalty_2019['Year'] = 2019
```

```python
penalty_2020 = pd.read_csv('Player_Penalties_2020.csv')
penalty_2020['Year'] = 2020
```

```python
penalty_2021 = pd.read_csv('Player_Penalties_2021.csv')
penalty_2021['Year'] = 2021
```

```python
penalty_2022 = pd.read_csv('Player_Penalties_2022.csv')
penalty_2022['Year'] = 2022
```

```python
penalty_2023 = pd.read_csv('Player_Penalties_2023.csv')
penalty_2023['Year'] = 2023
```

```python
penalty_2024 = pd.read_csv('Player_Penalties_2024.csv')
penalty_2024['Year'] = 2024
```

```python
total_penalties = pd.concat([penalty_2019, penalty_2020, penalty_2021, penalty_2022, penalty_2023, penalty_2024], axis=0)
```

```python
total_penalties
```

```python
total_penalties.to_csv("Total_Penalties.csv", index = False)
print("Data saved to Total_Penalties.csv")
```
