```python
import numpy as np
import pandas as pd
passing_2019 = pd.read_csv('Passing_Stats_2019.csv')
passing_2019['Year'] = 2019
```

```python
passing_2020 = pd.read_csv('Passing_Stats_2020.csv')
passing_2020['Year'] = 2020
```

```python
passing_2021 = pd.read_csv('Passing_Stats_2021.csv')
passing_2021['Year'] = 2021
```

```python
passing_2022 = pd.read_csv('Passing_Stats_2022.csv')
passing_2022['Year'] = 2022
```

```python
passing_2023 = pd.read_csv('Passing_Stats_2023.csv')
passing_2023['Year'] = 2023
```

```python
passing_2024 = pd.read_csv('Passing_Stats_2024.csv')
passing_2024['Year'] = 2024
```

```python
total_passing = pd.concat([passing_2019, passing_2020, passing_2021, passing_2022, passing_2023, passing_2024], axis=0)
```

```python
total_passing
```

```python
total_passing.to_csv("Total_Passing.csv", index = False)
print("Data saved to Total_Passing.csv")
```
